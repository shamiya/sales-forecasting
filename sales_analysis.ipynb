# **Predictive Analysis**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
# Load data
df = pd.read_csv('/content/drive/MyDrive/mypython/to_do.csv')

# Ensure 'Fiscal Year' is in numeric format
df['Fiscal Year'] = pd.to_numeric(df['Fiscal year'], errors='coerce')

# Aggregate sales data by year
trend_data = df.groupby('Fiscal Year')['GBP Equivalent'].sum().reset_index()

# Set 'Fiscal Year' as the index
trend_data.set_index('Fiscal Year', inplace=True)

# Visualize the trend
plt.figure(figsize=(10, 6))
sns.lineplot(data=trend_data, marker='o')
plt.title("Sales Trend Over the Years")
plt.xlabel("Fiscal Year")
plt.ylabel("Total GBP Equivalent")
plt.grid(True)
plt.show()
```

```python
# Fit ARIMA model
model = ARIMA(trend_data, order=(2,1,2))  # (p,d,q) values can be tuned
model_fit = model.fit()

# Forecast the next 3 years
forecast_years = [trend_data.index[-1] + i for i in range(1, 4)]
forecast = model_fit.forecast(steps=3)

# Convert forecast to DataFrame
forecast_df = pd.DataFrame({'Fiscal year': forecast_years, 'Forecasted Sales': forecast.values})
forecast_df.set_index('Fiscal year', inplace=True)
# Plot actual vs forecasted sales
plt.figure(figsize=(10, 6))

# Actual Sales
sns.lineplot(x=trend_data.index, y=trend_data['GBP Equivalent'], marker='o', label="Actual Sales")

# Forecasted Sales
sns.lineplot(x=forecast_df.index, y=forecast_df['Forecasted Sales'], marker='o', linestyle="dashed", label="Forecasted Sales")

plt.title("Sales Forecast")
plt.xlabel("Fiscal Year")
plt.ylabel("Total GBP Equivalent")
plt.legend()
plt.grid(True)
plt.show()

```
